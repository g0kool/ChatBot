```python
# importing necessary libraries
from transformers import AutoTokenizer
import gradio as gr
from transformers import pipeline
model = 'deepset/roberta-base-squad2' # deepset/roberta-base-squad2
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
nlp = pipeline("question-answering", model=model, tokenizer=tokenizer)

context = '''My name is gokool and I love working on the LLMs and Generative AI
and I live in Bangalore
'''
question = "what is my name?"
result = nlp(question = question, context=context)
result

question = "what do I love?"
result = nlp(question = question, context=context)
result

# creating the function for Gradio
def func(context, question):
 result = nlp(question = question, context=context)
 return result['answer']

# creating the Gradio Chatbot interface
app = gr.Interface(fn=func, inputs = ['textbox', 'text'], outputs = 'textbox', title = "kool's Bot 😊", theme = 'dark-grass', description = 'Input your context and question to get answers!')

# launching the Chatbot app
app.launch(inline=False,share = True)
```
